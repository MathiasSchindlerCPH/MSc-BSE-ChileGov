Create some measures per nodes and add them as attribute (for community detection,...)

```python
degree_dict = dict(G.degree(G.nodes())) #degree 
nx.set_node_attributes(G, degree_dict, 'degree')

degree_dict_in = dict(G.in_degree(G.nodes())) #in degree
nx.set_node_attributes(G, degree_dict_in, 'degree_in')

degree_dict_out = dict(G.out_degree(G.nodes())) #out degree
nx.set_node_attributes(G, degree_dict_out, 'degree_out')

eigenvector_dict = nx.eigenvector_centrality(G) # Run eigenvector centrality
nx.set_node_attributes(G, eigenvector_dict, 'eigenvector')

degree_dict = nx.degree_centrality(G) # Run degree centrality
nx.set_node_attributes(G, degree_dict, 'degree_cent')
```

```python
# Segregation for the appendix
def get_SSI(Gr,left_users,right_users,recip=True):
    G_und=Gr.to_undirected(reciprocal=recip)
    CC_list = [G_und.subgraph(c).copy() for c in nx.connected_components(G_und)]
    data = []
    
    for CC in CC_list:
        if CC.number_of_nodes() == 1:
            if str(list(CC.nodes)[0] )in left_users:
                affilitation = "Left"
                
            if str(list(CC.nodes)[0]) in right_users:
                affilitation = "Right"
            
            data.append({
                    'SSI_CC' : 0,
                    'Size_CC' : 1,
                    'Aff' : affilitation
                })
        
        else:
            try:
                # Get the normalized adjancency matrix
                mat = nx.adjacency_matrix(CC).todense() # adjancency matrix
                nor_mat = mat / np.tile(np.sum(mat, 1), (1, mat.shape[1])) # normalized

                # Convert the normalized adjancency matrix into a graph as the previous one (weighted this time)
                G_nm = nx.from_numpy_matrix(nor_mat, create_using=nx.DiGraph()) # Creation of the graph object
                G_nm = nx.relabel_nodes(G_nm, dict(zip(range(G_nm.number_of_nodes()), CC.nodes()))) # assignation of nodes ides

                for aff in ['Left','Right']:
                    # Over each race compute the SSI
                    # Keep nodes of the same race
                    if aff=="Left":
                        nodes_race=list(set(left_users) & set(list(CC.nodes)))
                        
                    elif aff=="Right":
                        nodes_race=list(set(right_users) & set(list(CC.nodes)))
                    
                    G_nm_w = G_nm.subgraph(nodes_race)

                    mat_r = nx.adjacency_matrix(G_nm_w).todense() # adjancency matrix
                    e_val, e_vec = np.linalg.eigh(mat_r) # Obtain eigenvalues and eigen vectors
                    pos = np.argmax(e_val) # Position of the largest eigenvalue and eigenvector
                    SSI = abs(e_val[pos]) # The largest eigenvalue is the SSI for the group
                    
                
                    data.append({
                            'SSI_CC' : SSI,
                            'Size_CC' : len(nodes_race),
                            'Aff' : aff
                         })
            except:
                pass

    data_frame=pd.DataFrame(data)
    data_frame['weight_SSI']=data_frame['SSI_CC']*data_frame['Size_CC']
    total_SSI=sum(data_frame['weight_SSI'])/sum(data_frame['Size_CC'])
    
    left_df=data_frame[data_frame['Aff']=="Left"]
    left_SSI=sum(left_df['weight_SSI'])/sum(left_df['Size_CC'])
    
    right_df=data_frame[data_frame['Aff']=="Right"]
    right_SSI=sum(right_df['weight_SSI'])/sum(right_df['Size_CC'])
    
    return data_frame, total_SSI, left_SSI, right_SSI
```
