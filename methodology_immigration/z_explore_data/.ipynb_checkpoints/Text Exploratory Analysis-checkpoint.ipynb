```python
tb=viz.top_bigrams(df,15,15," ")
```

```python
numb_of_users=[]
bigrams=[]

for cbig in tb:
    regex = '(?:^|\W)'+str(cbig[0][0]) +' '+str(cbig[0][1])+ '(?:$|\W)'
    df_big=pd.DataFrame()
    df_big=df.dropna(subset=['cleaned_text']).copy()
    df_big=df_big[df_big.cleaned_text.str.contains(regex)]
    df_big['count_big']=1
    df_1=df.merge(df_big[['id','count_big']],how="left",on="id")
    by_author=df_1.groupby(by='author.username').sum()
    by_author_used=by_author[by_author.count_big!=0]
    numb_of_users.append(by_author_used.shape[0])
    bigrams.append(cbig[0])

summary_big=pd.DataFrame()
summary_big['Bigram']=bigrams
summary_big['Number of users']=numb_of_users
```

```python
summary_big
```

```python
numb_of_users=[]
hashtags=[]
df=filters.f_affiliation(df_protests,'Left')

for h in thl:
    regex = '(?:^|\W)'+'#'+unidecode.unidecode(str(h[0])) + '(?:$|\W)'
    df_hash=pd.DataFrame()
    df_hash=df.dropna(subset=['cleaned_text']).copy()
    df_hash=df_hash[df_hash.cleaned_text.str.contains(regex)]
    df_hash['count_hasht']=1
    df_1=df.merge(df_hash[['id','count_hasht']],how="left",on="id")
    by_author=df_1.groupby(by='author.username').sum()
    by_author_used=by_author[by_author.count_hasht!=0]
    numb_of_users.append(by_author_used.shape[0])
    hashtags.append(h[0])

summary=pd.DataFrame()
summary['Hashtag']=hashtags
summary['Number of users']=numb_of_users

summary
```

```python
numb_of_users=[]
hashtags=[]
df=filters.f_affiliation(df_protests,'Right')

for h in thr:
    regex = '(?:^|\W)'+unidecode.unidecode(str(h[0])) + '(?:$|\W)'
    df_hash=pd.DataFrame()
    df_hash=df.dropna(subset=['tweet_hashtags']).copy()
    df_hash=df_hash[df_hash.tweet_hashtags.str.contains(regex)]
    df_hash['count_hasht']=1
    df_1=df.merge(df_hash[['id','count_hasht']],how="left",on="id")
    by_author=df_1.groupby(by='author.username').sum()
    by_author_used=by_author[by_author.count_hasht!=0]
    numb_of_users.append(by_author_used.shape[0])
    hashtags.append(h[0])

summary=pd.DataFrame()
summary['Hashtag']=hashtags
summary['Number of users']=numb_of_users

summary
```
